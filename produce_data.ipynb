```python
import pandas as pd
import numpy as np
from kafka import KafkaProducer # pip install kafka-python
from time import time, sleep
```

```python
df_gps = pd.read_csv('processed_jakarta_gps.csv')
```

```python
list_gps = df_gps.to_numpy().tolist()
print(list_gps[0])
```

```python
producer = KafkaProducer(bootstrap_servers='localhost:9092')
```

```python
for i in range(len(list_gps)):
    bus_code = list_gps[i][0]
    date_time = list_gps[i][2]
    longitude = list_gps[i][3]
    latitude = list_gps[i][4]
    speed = list_gps[i][5]
    
    # create CSV structure
    msg = f'{date_time},{bus_code},{longitude},{latitude},{speed}'

    # send to Kafka
    producer.send('weather', bytes(msg, encoding='utf8'))
    print(f'Bus({bus_code}) sending data to kafka')

    # count += 1
    sleep(0.001)
```
